```python
import pandas as pd
import numpy as np
print ("Hello Capstone Project Course!")
```

```python
Df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
Df = Df[Df.Borough != 'Not assigned']
Df.loc[Df['Neighbourhood'] == "Not assigned"] = Df['Borough']
Df.groupby(['Postcode','Borough'], sort = False).agg(lambda x: ', '.join(x))
```

```python
coordinates = pd.read_csv('https://cocl.us/Geospatial_data')
coordinates.rename(columns = {'Postal Code': 'Postcode'}, inplace = True)
coordinates
```

```python
Df1 = Df.merge(coordinates, on = 'Postcode')
```

```python
Df1.groupby(['Postcode', 'Borough', 'Latitude', 'Longitude'], sort = False).agg(lambda x: ', '.join(x))
```

```python
!pip install folium
import folium

#creating map
map_Toronto = folium.Map(location=[43.6523, -79.3832], zoom_start=10.7)

#adding markers
for Latitude, Longitude, Borough, Neighbourhood in zip(Df1['Latitude'], Df1['Longitude'], Df1['Borough'], Df1['Neighbourhood']):
    label = '{}, {}'.format(Neighbourhood, Borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [Latitude, Longitude],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7).add_to(map_Toronto)

map_Toronto
```

Most of the neighborhoods are clustered in the downtown area and then spread out very evenly.
