```python
Df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
Df = Df[Df.Borough != 'Not assigned']
Df.loc[Df['Neighbourhood'] == "Not assigned"] = Df['Borough']
Df.groupby(['Postcode','Borough'], sort = False).agg(lambda x: ', '.join(x))
```

```python
coordinates = pd.read_csv('https://cocl.us/Geospatial_data')
coordinates.rename(columns = {'Postal Code': 'Postcode'}, inplace = True)
coordinates
```

```python
Df1 = Df.merge(coordinates, on = 'Postcode')
```

```python
Df1.groupby(['Postcode', 'Borough', 'Latitude', 'Longitude'], sort = False).agg(lambda x: ', '.join(x))
```
